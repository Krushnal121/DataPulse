```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import networkx as nx
from wordcloud import WordCloud
import squarify
import numpy as np
import folium
from scipy.stats import gaussian_kde
from bokeh.plotting import figure, output_notebook, show
from bokeh.io import push_notebook
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool
from bokeh.layouts import gridplot
```

```python
# Load the dataset
df = pd.read_csv('household_power_consumption.txt', sep=';', parse_dates={'datetime': ['Date', 'Time']}, infer_datetime_format=True, na_values=['?'], low_memory=False)
```

```python
# Drop rows with missing values
df.dropna(inplace=True)
```

```python
# Convert columns to appropriate data types
df['Global_active_power'] = df['Global_active_power'].astype(float)
df['Global_reactive_power'] = df['Global_reactive_power'].astype(float)
df['Voltage'] = df['Voltage'].astype(float)
df['Global_intensity'] = df['Global_intensity'].astype(float)
df['Sub_metering_1'] = df['Sub_metering_1'].astype(float)
df['Sub_metering_2'] = df['Sub_metering_2'].astype(float)
df['Sub_metering_3'] = df['Sub_metering_3'].astype(float)
```

```python
# Set the index to datetime for time series analysis
df.set_index('datetime', inplace=True)

# Resample data to daily averages
df_daily = df.resample('D').mean()
```

```python
# Basic Charts
# 1. Bar Chart
df_daily['Global_active_power'].plot(kind='bar', figsize=(15, 5))
plt.title('Daily Global Active Power')
plt.ylabel('Global Active Power (kilowatts)')
plt.show()

# 2. Line Chart
df_daily['Global_active_power'].plot(kind='line', figsize=(15, 5))
plt.title('Daily Global Active Power')
plt.ylabel('Global Active Power (kilowatts)')
plt.show()

# 3. Pie Chart
df_daily['Global_active_power'].iloc[:7].plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Global Active Power Distribution (First 7 Days)')
plt.ylabel('')
plt.show()
```

```python
# 4. Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df_daily['Global_active_power'], df_daily['Voltage'], alpha=0.5)
plt.title('Global Active Power vs Voltage')
plt.xlabel('Global Active Power (kilowatts)')
plt.ylabel('Voltage (volts)')
plt.show()

# 5. Histogram
df['Global_active_power'].plot(kind='hist', bins=50, figsize=(10, 6))
plt.title('Histogram of Global Active Power')
plt.xlabel('Global Active Power (kilowatts)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Advanced Charts
# 6. Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Global_active_power', 'Global_reactive_power', 'Voltage']])
plt.title('Box Plot of Power Consumption and Voltage')
plt.show()

# 7. Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# 8. Area Chart
df_daily.plot.area(figsize=(15, 5), alpha=0.4)
plt.title('Daily Consumption and Voltage')
plt.ylabel('Value')
plt.show()

# 9. Sankey Diagram
fig = go.Figure(go.Sankey(
    node = dict(
        pad = 15,
        thickness = 20,
        line = dict(color = "black", width = 0.5),
        label = ["Global Active Power", "Global Reactive Power", "Sub Metering 1", "Sub Metering 2", "Sub Metering 3", "Others"],
        color = ["blue", "green", "red", "yellow", "purple", "orange"]
    ),
    link = dict(
        source = [0, 0, 0, 1, 1, 2, 3, 4],
        target = [2, 3, 4, 5, 2, 5, 5, 5],
        value = [8, 4, 2, 8, 4, 2, 2, 2]
    )))
fig.update_layout(title_text="Basic Sankey Diagram", font_size=10)
fig.show()
```

```python
# 10. Radar Chart (Spider Chart)
categories = ['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity']
values = df[categories].mean().values.flatten().tolist()
values += values[:1]
categories += categories[:1]
fig = go.Figure(data=go.Scatterpolar(r=values, theta=categories, fill='toself'))
fig.update_layout(polar=dict(radialaxis=dict(visible=True)), showlegend=False)
fig.show()

# 11. Network Graph
G = nx.Graph()
for i in range(10):
    G.add_node(i)
for i in range(10):
    for j in range(i+1, 10):
        G.add_edge(i, j)
plt.figure(figsize=(10, 6))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, edge_color='gray')
plt.title('Network Graph')
plt.show()
```

```python
# 12. Donut Chart
df_daily['Global_active_power'].iloc[:5].plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), wedgeprops=dict(width=0.3))
plt.title('Donut Chart of Global Active Power (First 5 Days)')
plt.ylabel('')
plt.show()

# 13. Gauge Chart
fig = go.Figure(go.Indicator(
    mode = "gauge+number",
    value = df_daily['Global_active_power'].mean(),
    title = {'text': "Average Global Active Power"},
    gauge = {'axis': {'range': [None, df_daily['Global_active_power'].max()]}}
))
fig.show()
```

```python
# 14. Sunburst Chart
fig = px.sunburst(df_daily.reset_index(), path=['datetime', 'Global_active_power'], values='Global_active_power')
fig.show()

# 15. Hexbin Plot
plt.figure(figsize=(10, 6))
plt.hexbin(df['Global_active_power'], df['Voltage'], gridsize=30, cmap='Blues')
plt.colorbar(label='count in bin')
plt.xlabel('Global Active Power')
plt.ylabel('Voltage')
plt.title('Hexbin Plot')
plt.show()
```

```python
# 16. Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Global_active_power', data=df)
plt.title('Violin Plot of Global Active Power')
plt.show()

# 17. Parallel Coordinates
from pandas.plotting import parallel_coordinates
plt.figure(figsize=(15, 6))
parallel_coordinates(df.reset_index(), 'datetime', cols=['Global_active_power', 'Global_reactive_power', 'Voltage'], color=('#556270', '#4ECDC4', '#C7F464'))
plt.title('Parallel Coordinates Plot')
plt.show()
```

```python
# 18. Choropleth Map
# For demonstration, we will use a dataset with geographical data
map_data = pd.DataFrame({
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Latitude': [40.7128, 34.0522, 41.8781, 29.7604, 33.4484],
    'Longitude': [-74.0060, -118.2437, -87.6298, -95.3698, -112.0740],
    'Global_active_power': [20, 30, 25, 35, 40]
})

fig = px.scatter_geo(map_data, lat='Latitude', lon='Longitude', color='Global_active_power',
                     size='Global_active_power', hover_name='City', projection='natural earth')
fig.update_layout(title='Choropleth Map of Power Consumption')
fig.show()
```

```python
# 19. Strip Plot
plt.figure(figsize=(10, 6))
sns.stripplot(x='Global_active_power', data=df)
plt.title('Strip Plot of Global Active Power')
plt.show()

# 20. Swarm Plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Global_active_power', data=df)
plt.title('Swarm Plot of Global Active Power')
plt.show()
```
